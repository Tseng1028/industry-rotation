```python
import pandas as pd
import matplotlib.pyplot as plt

# 假設你的資料存放在一個 CSV 檔案中
# 讀取資料，請將 'your_file.csv' 替換為你的檔案名稱
rolling_predictions = pd.read_csv(r'C:\Users\User\Desktop\quant\Sector Allocation Framework under Reinforcement Learning Rotation\industry-rotation\industryRL_v3\rolling_predictions4.csv')
```

```python
rolling_predictions#[rolling_predictions['prediction_accuracy'] == True]
```

```python
# 計算累積回報
rolling_predictions['cumulative_return'] = rolling_predictions['predicted_return'].cumsum()

# 繪製累積回報圖
plt.figure(figsize=(10, 6))
plt.plot(rolling_predictions['cumulative_return'], marker='o', label='Cumulative Return')
plt.title('Cumulative Predicted Return')
plt.xlabel('Observation')
plt.ylabel('Cumulative Return')
plt.grid()
plt.legend()
plt.show()
```

```python


# 設置圖片大小和風格
plt.figure(figsize=(10, 6))

# 使用 predict_end 作為 x 軸
plt.plot(pd.to_datetime(rolling_predictions['predict_end']), 
         rolling_predictions['cumulative_return'], 
         marker='o', 
         label='Cumulative Return')

# 設置標題和標籤
plt.title('Cumulative Predicted Return')
plt.xlabel('Prediction End Date')
plt.ylabel('Cumulative Return')

# 調整 x 軸日期標籤的角度，避免重疊
plt.xticks(rotation=45)

# 自動調整圖表佈局，確保所有元素都能顯示
plt.tight_layout()

# 添加網格和圖例
plt.grid()
plt.legend()

plt.show()
```
